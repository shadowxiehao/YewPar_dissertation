```python
import matplotlib.pyplot as plt
import numpy as np

# Localities
localities = [20, 15, 10, 5]

# Performance-Driven Workstealing Policy times
pdw_times = [
    [14829, 12791, 16863, 17266, 16410, 13920, 14489, 17646, 21343, 12549],
    [21533, 19751, 20677, 25336, 21294, 25132, 18389, 17850, 19703, 16009],
    [24463, 27543, 25995, 32465, 32764, 25782, 27933, 22756, 32031, 23569],
    [51414, 64284, 58569, 66468, 65849, 55616, 62883, 64843, 63617, 60311]
]

# Depth Pool Workstealing Policy times
dpw_times = [
    [16893, 15822, 17997, 15739, 16006, 13570, 17827, 17189, 16246, 12595],
    [21678, 22525, 19146, 18976, 20766, 22809, 19951, 20347, 21012, 19747],
    [29963, 28186, 28006, 32229, 27420, 29927, 29617, 28761, 28925, 26073],
    [79945, 50156, 51396, 53098, 99086, 99490, 75471, 51503, 64685, 68091]
]

# Calculate the average times for each policy
avg_pdw_times = [np.mean(times) for times in pdw_times]
avg_dpw_times = [np.mean(times) for times in dpw_times]

# Calculate speedup percentages
speedup_percentages = [(dpw - pdw) / dpw * 100 for dpw, pdw in zip(avg_dpw_times, avg_pdw_times)]

# Adjusting the titles, colors, and the y-axis scaling for the average time graph

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10))

# Plotting the time difference for each locality with a logarithmic scale
ax1.plot(localities, avg_pdw_times, marker='o', label='Performance-Driven', color='dodgerblue')
ax1.plot(localities, avg_dpw_times, marker='o', label='Depth Pool', color='coral')
ax1.set_yscale('log')
ax1.set_xlabel('Localities')
ax1.set_ylabel('Average Time (ms, log scale)')
ax1.set_title('Average Time Comparison for maxclique-16')
ax1.legend()
ax1.grid(True, which="both", ls="--", c='0.65')

# Plotting the speedup percentages
bars = ax2.bar(localities, speedup_percentages, color=['limegreen' if percent > 0 else 'tomato' for percent in speedup_percentages])
ax2.plot(localities, speedup_percentages, marker='o', color='black', label='Percentage Change')
ax2.set_xlabel('Localities')
ax2.set_ylabel('Speedup Percentage (%)')
ax2.set_title('Speedup Percentage using Performance-Driven for maxclique-16')
ax2.legend()
ax2.grid(True, axis='y', ls="--", c='0.65')

# Annotating the bars with their respective percentages
for bar in bars:
    yval = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom', color='black')

plt.tight_layout()
#plt.savefig("../image/time_comparison_max_solo.pdf")
plt.show()
```
